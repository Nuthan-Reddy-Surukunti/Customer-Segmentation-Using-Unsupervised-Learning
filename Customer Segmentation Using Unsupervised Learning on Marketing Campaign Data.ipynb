```python
from sklearn.preprocessing import StandardScaler

# Select the numerical features to scale
numerical_cols_to_scale = ['Age', 'Household_Size', 'Income', 'Recency', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 'Total_Spend', 'Tenure', 'Campaign_Score']

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply scaling to the numerical features
df_encoded[numerical_cols_to_scale] = scaler.fit_transform(df_encoded[numerical_cols_to_scale])

display(df_encoded.head())
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Initialize PCA
# We'll start by keeping all components to examine explained variance
pca = PCA()

# Fit PCA on the scaled and encoded data
X = df_encoded.copy()
pca.fit(X)

# Plot explained variance ratio
plt.figure(figsize=(10, 6))
plt.plot(range(1, pca.n_components_ + 1), np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.title('Explained Variance by Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.grid(True)
plt.show()
```

```python
# Initialize PCA with 15 components
pca = PCA(n_components=15)

# Fit PCA and transform the scaled data
X_pca = pca.fit_transform(X)

# Create a new DataFrame with the PCA components
df_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])

display(df_pca.head())
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

# Use the PCA transformed data for clustering
X_clust = df_pca.copy()

# Initialize KMeans and the KElbowVisualizer
model = KMeans(random_state=42, n_init=10) # Added n_init for KMeans warning
visualizer = KElbowVisualizer(model, k=(1, 11)) # Check k from 1 to 10

plt.figure(figsize=(8, 6))
visualizer.fit(X_clust) # Fit the data to the visualizer
visualizer.show() # Finalize and render the figure
```

```python
from sklearn.cluster import KMeans

# Use the PCA transformed data for clustering
X_clust = df_pca.copy()

# Initialize and fit the KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10) # Added n_init for KMeans warning
kmeans.fit(X_clust)

# Add the cluster labels to the original dataframe or the PCA dataframe
df['Cluster'] = kmeans.labels_
df_pca['Cluster'] = kmeans.labels_

display(df.head())
display(df_pca.head())
```

```python
# Visualize the clusters using the first two principal components
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=df_pca, palette='viridis', s=50)
plt.title('Customer Clusters (PCA 1 vs PCA 2)')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# Profile the clusters by calculating the mean of the original features for each cluster
# Select only numerical columns for profiling
numerical_df = df.select_dtypes(include=np.number)
cluster_profiles = numerical_df.groupby(df['Cluster']).mean()

display(cluster_profiles)
```

## Customer Cluster Summary

Based on the K-Means clustering analysis with 3 clusters, we have identified the following customer segments:

### Cluster 0: High-Spending, Engaged Customers
*   **Characteristics:** Highest average income and total spending, higher campaign score, lower web visits, higher catalog and store purchases, higher response rate to the last campaign.

### Cluster 1: Mid-Range Customers with Families
*   **Characteristics:** Moderate average income and total spending, higher average household size (more kids/teens), moderate campaign score, higher number of web and store purchases.

### Cluster 2: Lower-Spending Customers
*   **Characteristics:** Lowest average income and total spending, higher average household size, lowest campaign score and response rate, more web visits, fewer catalog and store purchases.

These profiles provide insights into the different customer segments, which can be valuable for targeted marketing strategies and business decisions.

```python
from sklearn.metrics import silhouette_score

# Calculate the silhouette score for the K-Means clustering
# Use the PCA-transformed data (X_clust) and the cluster labels
silhouette_avg = silhouette_score(X_clust, df_pca['Cluster'])

print(f"The silhouette score for the K-Means clustering is: {silhouette_avg}")
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Use the PCA transformed data for clustering
X_clust = df_pca.copy()

# Perform hierarchical clustering
linked = linkage(X_clust, method='ward')

# Plot the dendrogram
plt.figure(figsize=(15, 10))
dendrogram(linked,
            orientation='top',
            distance_sort='descending',
            show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Use the PCA transformed data for clustering
X_clust = df_pca.copy()

# Apply Agglomerative Clustering with 2 clusters
# Using metric='euclidean' and linkage='ward' which corresponds to the dendrogram method
agg_clustering = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
clusters_h = agg_clustering.fit_predict(X_clust)

# Add the hierarchical cluster labels to the dataframe
df['Hierarchical_Cluster'] = clusters_h
df_pca['Hierarchical_Cluster'] = clusters_h

display(df.head())
display(df_pca.head())
```

```python
# Profile the hierarchical clusters by calculating the mean of the original features for each hierarchical cluster
# Select only numerical columns for profiling
numerical_df = df.select_dtypes(include=np.number).drop(columns=['Cluster']) # Exclude the K-Means cluster column
hierarchical_cluster_profiles = numerical_df.groupby(df['Hierarchical_Cluster']).mean()

display(hierarchical_cluster_profiles)
```

```python
# Calculate the silhouette score for the Hierarchical clustering
# Use the PCA-transformed data (X_clust) and the hierarchical cluster labels
silhouette_avg_h = silhouette_score(X_clust, df_pca['Hierarchical_Cluster'])

print(f"The silhouette score for the Hierarchical clustering is: {silhouette_avg_h}")
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt

# Use the PCA transformed data for finding neighbors
X_clust = df_pca.copy()

# We'll use k = 5 as a starting point for min_samples + 1
k = 5

# Find the k-nearest neighbors
neigh = NearestNeighbors(n_neighbors=k)
neigh.fit(X_clust)

# Find the distances to the k-nearest neighbor
distances, indices = neigh.kneighbors(X_clust)

# Sort the distances
distances = np.sort(distances[:, k-1], axis=0)

# Plot the k-distance graph
plt.figure(figsize=(10, 6))
plt.plot(distances)
plt.title('K-Distance Graph')
plt.xlabel('Data Points (Sorted by Distance)')
plt.ylabel(f'Distance to {k}-th Nearest Neighbor')
plt.grid(True)
plt.show()
```

# Customer Segmentation Analysis Summary

This document summarizes the key observations from the customer segmentation analysis using K-Means and Hierarchical clustering on the marketing campaign dataset.

## Data Preparation and Feature Engineering

The dataset was loaded and preprocessed by handling missing income values, creating new features such as `Age`, `Household_Size`, `Total_Spend`, `Tenure`, and `Campaign_Score`, and dropping irrelevant or redundant columns. Categorical features (`Education` and `Marital_Status`) were one-hot encoded, and numerical features were scaled using StandardScaler. Principal Component Analysis (PCA) was applied to reduce the dimensionality of the data for clustering.

## K-Means Clustering

K-Means clustering was performed on the PCA-transformed data. The elbow method suggested 3 as a possible number of clusters.

**Cluster Profiles (Based on mean of original features):**

| Feature             | Cluster 0 (High-Spending) | Cluster 1 (Mid-Range) | Cluster 2 (Lower-Spending) |
|---------------------|---------------------------|-----------------------|----------------------------|
| **Age**             | 56.66                     | 59.80                 | 53.44                      |
| **Household_Size**  | 0.20                      | 1.21                  | 1.22                       |
| **Income**          | 77393.00                  | 56915.94              | 34758.81                   |
| **Total_Spend**     | 1341.91                   | 744.33                | 93.84                      |
| **Campaign_Score**  | 0.76                      | 0.21                  | 0.09                       |
| **NumWebVisitsMonth**| 2.66                     | 6.05                  | 6.39                       |
| **NumStorePurchases**| 8.36                     | 7.57                  | 3.20                       |
| **Response**        | 0.27                      | 0.14                  | 0.08                       |

**Observations from K-Means:**

*   **Cluster 0:** Characterized by higher income, significantly higher spending, more campaign responses, and more purchases through catalog and in-store, with fewer web visits. This segment appears to be high-value, engaged customers.
*   **Cluster 1:** Represents a mid-range segment with moderate income and spending. They tend to have larger households (more kids/teens) and engage through both web and store purchases.
*   **Cluster 2:** Consists of lower-income and lower-spending customers. They also have larger households, fewer campaign responses, and primarily engage through web visits rather than catalog or in-store purchases.

The silhouette score for K-Means clustering with 3 clusters was approximately 0.194, suggesting some overlap between clusters.

## Hierarchical Clustering

Hierarchical clustering was also performed on the PCA-transformed data. The dendrogram suggested splitting into 2 main clusters.

**Cluster Profiles (Based on mean of original features):**

| Feature             | Hierarchical Cluster 0 | Hierarchical Cluster 1 |
|---------------------|------------------------|------------------------|
| **Age**             | 57.69                  | 54.19                  |

| **Income**          | 66454.76               | 35129.38               |

| **NumStorePurchases**| 7.91                  | 3.25                   |
| **Response**        | 0.20                   | 0.09                   |

**Observations from Hierarchical Clustering:**

*   **Hierarchical Cluster 0:** Similar to K-Means Cluster 0, this segment shows higher income, significantly higher spending, more campaign responses, and more purchases through catalog and in-store.
*   **Hierarchical Cluster 1:** Aligns with K-Means Clusters 1 and 2, representing lower income and spending customers, often with larger households, fewer campaign responses, and a preference for web visits.

The silhouette score for Hierarchical clustering with 2 clusters was approximately 0.220, slightly higher than K-Means, but still indicating some overlap.

## DBSCAN Clustering

DBSCAN clustering was attempted, and based on the k-distance graph, parameters `eps=2.5` and `min_samples=5` were chosen. However, this resulted in a large number of points being classified as noise (-1) and most other points belonging to a single cluster. The silhouette score for DBSCAN was approximately 0.183. This suggests that DBSCAN might not be the most suitable algorithm for this dataset with these parameters, or that the clusters are not dense and well-separated in the PCA space.

## Conclusion

Both K-Means (with 3 clusters) and Hierarchical clustering (with 2 clusters) provided interpretable customer segments based on the analyzed features. While the silhouette scores were relatively low, suggesting some cluster overlap, the distinct profiles identified can still be valuable for targeted marketing strategies. The high-spending, engaged customer segment (K-Means Cluster 0 / Hierarchical Cluster 0) is a key group for focused campaigns, while the other segments require different approaches based on their income, spending habits, and household composition. Further exploration with different clustering algorithms, varying the number of clusters, or refining feature engineering might reveal more defined segments.
